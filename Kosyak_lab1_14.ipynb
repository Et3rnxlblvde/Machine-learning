Створіть одновимірний масив (вектор) із геометричною прогресією, де кожне
наступне число менше попереднього в 3 рази, починаючи з 81, та виведіть його
значення

```python
import numpy as np

start_value = 81
ratio = 1/3
num_elements = 6

geometric_progression = start_value * ratio**np.arange(num_elements)

print("Масив із геометричною прогресією:")
print(geometric_progression)
```

Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його оберненими
значеннями дробових частин від ділення кожного числа на 2, починаючи з 1, та
виведіть його значення.

```python
import numpy as np

matrix = np.zeros((3, 3))

for i in range(3):
    for j in range(3):
        num = i * 3 + j + 1
        fraction = (num / 2) - (num // 2)
        if fraction != 0:
            matrix[i, j] = 1 / fraction
        else:
            matrix[i, j] = 0

print(matrix)
```

Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в
діапазоні від 0 до 1 та знайдіть середнє арифметичне елементів, розташованих на
побічній діагоналі.

```python
import numpy as np

array = np.random.rand(5, 5)

side_diagonal_elements = [array[i, 4 - i] for i in range(5)]

mean_side_diagonal = np.mean(side_diagonal_elements)

print("Масив:")
print(array)
print("Елементи на побічній діагоналі:", side_diagonal_elements)
print("Середнє арифметичне елементів на побічній діагоналі:", mean_side_diagonal)
```

Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні
від 1 до 10 та знайдіть її обернену матрицю.

```python
import numpy as np

matrix = np.random.randint(1, 11, size=(3, 3))

if np.linalg.det(matrix) == 0:
    print("Матриця не є оберненою (визначник дорівнює нулю).")
else:
    inverse_matrix = np.linalg.inv(matrix)
    print("Матриця:")
    print(matrix)
    print("Обернена матриця:")
    print(inverse_matrix)
```

Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні
від 1 до 10 та знайдіть добуток елементів матриці, які менше за їх середнє
арифметичне значення.

```python
import numpy as np

matrix = np.random.randint(1, 11, size=(3, 3))

mean_value = np.mean(matrix)

elements_less_than_mean = matrix[matrix < mean_value]

if len(elements_less_than_mean) > 0:
    product = np.prod(elements_less_than_mean)
else:
    product = None

print("Матриця (3x3):")
print(matrix)
print("Середнє арифметичне:", mean_value)
print("Елементи менше середнього арифметичного:", elements_less_than_mean)
print("Добуток елементів менше середнього арифметичного:", product)
```

Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в
діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить найбільше ціле значення
в кожному рядку матриці

```python
import numpy as np

matrix = np.random.uniform(-1, 1, size=(3, 3))

def max_integer_in_row(row):
    integers = np.floor(row).astype(int)
    return np.max(integers)

max_integers = np.array([max_integer_in_row(row) for row in matrix])

print("Матриця (3x3):")
print(matrix)
print("Вектор-стовпчик з найбільшими цілими значеннями в кожному рядку:")
print(max_integers)
```
